# fMRIprep

What does fMRIprep do?
- Motion correction
- Slice-timing correction
- Spatical normalization
- Confound (mostly motion) regressor estimation
- Basic anatomical image pre-processing (bias field correction, segmentation, spatial normalization, etc.)

What does fMRIprep *not* do?
- Spatial smoothing
- Termporal filtering
- Confound regression
- Signal scaling

## Installation

- Recommand way: docker (stable, consistent, free of dependencies)
- No easy way to run docker on HPC cluster like Talapas (permission issue)
- Use singularity
    - Designed for HPC
    - Already configed on Talapas
    - Very easy to use

### How to build fMRIprep singularity image on Mac?
- Install [docker desktop](https://hub.docker.com/editions/community/docker-ce-desktop-mac?utm_location=header)
- Pull fMRIprep's offical docker image and convert it to singularity format  
`docker run --privileged -t --rm --mount type=volume,source=singularity_cache,target=/singularity_cache --mount type=bind,source=/var/run/docker.sock,target=/var/run/docker.sock --mount type=bind,source=~/Softwares/singularity_images,target=/images -e "SINGULARITY_CACHEDIR=/singularity_cache" -e "SINGULARITY_TMPDIR=/singularity_cache" quay.io/singularity/singularity:v3.7.4 build /images/fmriprep-v21.0.1.sif docker://nipreps/fmriprep:21.0.1`  
- Upload singularity image to Talapas

### Which version should I use?
- Use the latest one when possible
- Consistent within a project (same major and minor version, e.g., 21.0.X)
- Be aware of patches ([fMRIprep changelog](https://github.com/nipreps/fmriprep/blob/master/CHANGES.rst))

## Run fMRIprep on Talapas

Prepare singularity : Add `module load singularity` in *~/.bashrc* file (only need to do it once)

Submit fMRIprep task as SLURM job
`sbatch run_fmriprep.sbatch`

Example sbatch job file:

```
#!/bin/bash

#SBATCH --account=kuhl_lab  
#SBATCH --partition=kuhl  
#SBATCH --job-name=fMRIprep  
#SBATCH --nodes=1  
#SBATCH --cpus-per-task=8  
#SBATCH --mem=24GB  
#SBATCH --time=1-00:00:00  
#SBATCH --output=%x_%A_%a.log  
#SBATCH --array=0-2

ARRAY_VALUE=(001 002 003)  
SUB_ID=${ARRAY_VALUE[${SLURM_ARRAY_TASK_ID}]}  
SINGULARITYENV_MKL_NUM_THREADS=8  
SINGULARITYENV_OPENBLAS_NUM_THREADS=8  
SINGULARITYENV_OMP_NUM_THREADS=8  
SINGULARITYENV_VECLIB_MAXIMUM_THREADS=8  
SINGULARITYENV_NUMEXPR_NUM_THREADS=8  
singularity run -e -B ${PROJECT_DIR}/bidsdata:/data,${OUTPUT_DIR}:/output,${TMP_DIR}:/works,${LICENSE_DIR}/license.txt:/license.txt ${CONTAINER_DIR}/fmriprep.sif /data /output/fmriprep participant --participant-label ${SUB_ID} --nprocs 8 --omp-nthreads 8 --mem_mb 24GB --output-spaces MNI152NLin6Asym:res-2 MNI152NLin2009cAsym T1w --slice-time-ref 0.5 --fd-spike-threshold 0.5 --fs-license-file /license.txt --fs-subjects-dir /output/freesurfer -w /works --notrack --stop-on-first-crash
```

Some decision points and fMRIprep arguments:
- Output space:
    - T1w (native resolution)
    - MNI152NLin2009cAsym (native resolution)
    - MNI152NLin6Asym:res-2 (2mm standard resolution 91x109x91)
- Precomputed FreeSurfer surfaces:
    - For volumetric, not a big deal
    - For surface-based analysis, I prefer run FreeSurfer separately and correct results manually (time consuming, but usually more accurate)
- CIFTI output
  - HCP like CIFTI file 
  - Add argument `--cifti-output 91k`
  - Different registration, sampling method compare to HCP pipeline
  - The result CIFTI file is valid, and compatible with normal HCP files.
- Slice time reference
  - Default: averaged slice time (the midpoint of a volume)
  - Argument `--slice-time-ref`: The time of the reference slice to correct BOLD values to, as a fraction acquisition time. 0 indicates the start, 0.5 the midpoint, and 1 the end of acquisition. The alias start corresponds to 0, and middle to 0.5. The default value is 0.5.
  - [Different GLM softwares have different default, adjust accordingly when build design matrix](https://reproducibility.stanford.edu/slice-timing-correction-in-fmriprep-and-linear-modeling). (e.g., FSL, midpoint 0.5; AFNI and nilearn: start point 0)

### Work with CIFTI file

CIFTI file is a format mixing surface and volume data representation. It usually consists 3 parts: left surface, right surface and volume (for subcortical regions). 
![](https://d3i71xaburhd42.cloudfront.net/3a598b14f5fad52f185f2ebfed969b1d1a199a91/31-Figure3-1.png)


````python
def get_vol_img_from_cifti(cifti_data: np.ndarray, axis: nib.cifti2.BrainModelAxis) -> nib.Nifti1Image:
    """
    Extract volume data as a volume image from CIFTI dscalar or dtseries file.

    Expect 'data' is a 2d array with BrainModel in the 2nd dim (dim=1)
    """

    assert isinstance(axis, nib.cifti2.BrainModelAxis)

    # Find volume voxels
    vol_mask = axis.volume_mask
    vox_indices = tuple(axis.voxel[vol_mask].T)  # ([x0, x1, ...], [y0, ...], [z0, ...])
    # Extract data from volume voxels and make a 4d/3d array
    cifti_data = cifti_data.T[vol_mask]  # Assume brainmodels axis is last, move it to front
    vol_data = np.zeros(axis.volume_shape + cifti_data.shape[1:], dtype=cifti_data.dtype)
    vol_data[vox_indices] = cifti_data  # "Fancy indexing"

    return nib.Nifti1Image(vol_data, axis.affine)


def get_surf_data_from_cifti(cifti_data: np.ndarray, axis: nib.cifti2.BrainModelAxis, surf_name: str) -> np.ndarray:
    """
    Extract surface data as a numpy array from CIFTI scalar or timeseries file.

    Expect 'data' is a 2d array with BrainModel in the 2nd dim (dim=1)
    """

    assert isinstance(axis, nib.cifti2.BrainModelAxis)
    if surf_name not in ["CIFTI_STRUCTURE_CORTEX_LEFT", "CIFTI_STRUCTURE_CORTEX_RIGHT"]:
        raise ValueError(
            f"Surface name {surf_name} is not supported.\n"
            "Valid values: CIFTI_STRUCTURE_CORTEX_LEFT, CIFTI_STRUCTURE_CORTEX_RIGHT"
        )

    # Loop through brain structures
    # Surface name is CIFTI_STRUCTURE_CORTEX_LEFT or CIFTI_STRUCTURE_CORTEX_RIGHT
    for name, data_indices, model in axis.iter_structures():
        if name == surf_name:
            # Find vertex
            vtx_indices = model.vertex
            # Extract data from surface vertices and make a 2d/1d array
            # assume brainmodels axis is last, move it to front
            cifti_data = cifti_data.T[data_indices]
            surf_data = np.zeros((vtx_indices.max() + 1,) + cifti_data.shape[1:], dtype=cifti_data.dtype)
            surf_data[vtx_indices] = cifti_data

            # Returned data has vertices index in the 1st dim (dim=0)
            return surf_data


def decompose_cifti_dscalar(img: nib.cifti2.Cifti2Image, dtype=np.float32) -> dict:
    """
    Split a CIFTI dscalar image into surface and volume parts.
    """

    data = {"SurfaceL": None, "SurfaceR": None, "Volume": None}
    # Read data in CIFTI file
    data_all = img.get_fdata(dtype=np.float64).astype(dtype)
    # Get BrainModel axis (assume the last dim for dscalar or dtseries)
    axis_bm = img.header.get_axis(1)
    # Split data for each brain structure
    if "CIFTI_STRUCTURE_CORTEX_LEFT" in axis_bm.name:
        data["SurfaceL"] = get_surf_data_from_cifti(data_all, axis_bm, "CIFTI_STRUCTURE_CORTEX_LEFT")
    if "CIFTI_STRUCTURE_CORTEX_RIGHT" in axis_bm.name:
        data["SurfaceR"] = get_surf_data_from_cifti(data_all, axis_bm, "CIFTI_STRUCTURE_CORTEX_RIGHT")
    if axis_bm.volume_shape is not None:
        data["Volume"] = get_vol_img_from_cifti(data_all, axis_bm)

    return data
````

````python
cifti_file = "/gpfs/projects/kuhl_lab/zhifangy/Project_Dipper/data/derivatives/singletrial_response/sub-001/sub-001_task-ers_run-1_space-fsLR_den-32k_desc-sm0pt0_beta.dscalar.nii"
cifti_data = decompose_cifti_dscalar(nib.load(cifti_file))
````
