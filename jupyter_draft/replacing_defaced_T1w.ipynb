```python
import os
import sys
from glob import glob
import shutil
```

```python
defaced = os.path.join('/gpfs/projects/kuhl_lab/wanjiag/', 'MONSTERA_defaced_T1', '*T1w__defaced__msk.nii.gz')
output = os.path.join('/gpfs/projects/kuhl_lab/wanjiag/', 'MONSTERA')
```

```python
defaced_files = glob(defaced)
```

```python
defaced_files.sort()
```

```python
defaced_files
```

```python
for f in defaced_files:
    f_name = f.split('/')[-1]
    sub_name = f_name.split('_')[0]
    new_f_name = f_name.split('__')[0]+'.nii.gz'
    output_path = os.path.join(output, sub_name, 'anat', new_f_name)
    
    try:
        os.remove(output_path)
        shutil.move(f, output_path)
        print(f"File moved successfully for {sub_name} at {output_path}")
    except FileNotFoundError:
        print(f"Error: {sub_name} Source file {source_path} not found.")
    except PermissionError:
        print(f"Error: {sub_name} Permission denied to move {source_path}.")
    except Exception as e:
        print(f"{sub_name} An unexpected error occurred: {e}")
```
